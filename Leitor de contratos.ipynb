```python
import pdfplumber
import re
from docx import Document
```

```python
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text

def extract_text_from_word(docx_path):
    doc = Document(docx_path)
    text = ""
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

def extract_values_with_context(file_path, file_type):
    if file_type == 'pdf':
        text = extract_text_from_pdf(file_path)
    elif file_type == 'docx':
        text = extract_text_from_word(file_path)
    else:
        raise ValueError("Unsupported file type. Use 'pdf' or 'docx'.")

    # Expressões regulares para encontrar datas, valores e palavras-chave
    date_pattern = r"\d{2}/\d{2}/\d{4}"  # Formato dia/mês/ano
    date_pattern_dmy = r"\d{2}-\d{2}-\d{4}"  # Formato dia-mês-ano
    value_pattern = r"R\$\s*\d+,\d{2}"
    keyword_patterns = [r"serviços", r"meses", r"prestação", r"mensalidade", 'data', 'vigência', 'janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
                        'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro', 'automático', 'expressa', 'tácita']

    dates = re.findall(date_pattern, text) + re.findall(date_pattern_dmy, text)
    values = re.findall(value_pattern, text)

    # Capturar as frases próximas às datas, valores e palavras-chave
    context = {}
    for date in dates:
        date_index = text.find(date)
        context[date] = text[max(0, date_index - 50):date_index + 50]

    for value in values:
        value_index = text.find(value)
        context[value] = text[max(0, value_index - 50):value_index + 50]

    for keyword_pattern in keyword_patterns:
        keywords = re.findall(keyword_pattern, text, flags=re.IGNORECASE)
        for keyword in keywords:
            keyword_index = text.find(keyword)
            context[keyword] = text[max(0, keyword_index - 50):keyword_index + 50]

    return context

# Caminho para o arquivo
file_path = r"C:\Users\Luis.Silva\Downloads\Anotações\Contratos Medilar\TERMO DE CONTRATAÇÃO - DDG 0800 055 8885.docm.pdf"
file_type = "pdf"  # ou "docx" para arquivos Word

# Extrair datas, valores e contexto do arquivo
context_dict = extract_values_with_context(file_path, file_type)

# Exibir os resultados organizados
for value, context in context_dict.items():
    print(f"{value}:\n{context}\n\n")
```
